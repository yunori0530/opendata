```python
def dateparse(dt):
    if (isinstance(dt, str)):
        try:
            regex = r'(\d+)年(\d+)月(\d+)日\s?(\d+)時(\d+)分(\d+)秒'
            matches = re.findall(regex, dt)
            second = int(matches[0][5])
            if (second > 59): 
                second = 59
            date = datetime(int(matches[0][0])+1911, int(matches[0][1]), int(matches[0][2]),\
                            int(matches[0][3]), int(matches[0][4]), second, 0)
        except:
            #print(dt)
            date = "2020-1-1"
    return date

url = "https://data.moi.gov.tw/MoiOD/System/DownloadFile.aspx?DATA=219560DB-5617-4C5F-9754-8A4B96AC5F42"
df = pd.read_csv(url, sep=',', header=0, parse_dates={'time':[0]}, date_parser=dateparse)
df.drop(df.tail(2).index,inplace=True)
df.info()
```

```python
# 提示 : 使用字串 contains 方法可比對所需之字串
mask1 = df["發生地點"].str.contains("臺南市")     # 發生地點 -- 臺南市麻豆區
mask2 = df["死亡受傷人數"].str.contains("受傷1")  # 死亡受傷人數 -- 受傷 1 人
mask3 = df["車種"].str.contains("自用-小客車")    # 車種 -- 自用-小客車
res = df[mask1 & mask2 & mask3]  
display(res)
```

```python
# 題目請執行這一 cell
def get_listitem(thelist):
  return random.choice(thelist)
places = df['發生地點'].values
counties = []
regions = []
crs = []
pattern = r"^(\w{1,2}?[縣市])(\w{1,3}?[鄉|鎮|市|區])"
prog = re.compile(pattern)
for p in places:
    result = prog.match(p)
    g = result.groups()
    if g[0] not in counties:
        counties.append(g[0])
    if g[1] not in regions:
        regions.append(g[1])
    if [g[0],g[1]] not in crs:
        crs.append([g[0],g[1]])
```
